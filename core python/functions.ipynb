```python
#Q2.  *args + RECURSION: Write a recursive function that takes *args of numbers and returns their sum WITHOUT using the built-in sum().
def recursive_sum(*args):
    if len(args)==0:
        return 0
    if len(args)==1:
        return args[0]
    return args[0]+recursive_sum(*args[1:])
print(recursive_sum(10,20,30,40))
```
